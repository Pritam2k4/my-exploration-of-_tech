```python
import matplotlib.pyplot as plt
import numpy as np

# Define x range
x = np.linspace(-10, 10, 400)

# Functions
y_linear = 2 * x + 3
y_quadratic = x**2 - 4
y_cubic = x**3 - 3*x
y_exp = np.exp(0.3 * x)
y_log = np.log(x + 11)  # shifted right to avoid log(0)
y_trig = np.sin(x)

# Plotting function
def plot_function(x, y, title):
    plt.figure(figsize=(5,4))
    plt.plot(x, y, label=title)
    plt.axhline(0, color='black', linewidth=0.7)
    plt.axvline(0, color='black', linewidth=0.7)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Generate plots
plot_function(x, y_linear, "Linear Function: y = 2x + 3")
plot_function(x, y_quadratic, "Quadratic Function: y = x² - 4")
plot_function(x, y_cubic, "Cubic Function: y = x³ - 3x")
plot_function(x, y_exp, "Exponential Function: y = e^(0.3x)")
plot_function(x, y_log, "Logarithmic Function: y = log(x+11)")
plot_function(x, y_trig, "Trigonometric Function: y = sin(x)")
```
